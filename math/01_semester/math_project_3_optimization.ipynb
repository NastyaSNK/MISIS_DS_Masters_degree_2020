```python
import cmath

def zero_grad(a, b, c, d):
    desc = 4*b**2-12*a*c
    if desc >= 0:
        root_1 = (-2*b+desc**0.5)/(6*a)
        root_2 = (-2*b-desc**0.5)/(6*a)
    if desc==0:
        return sorted([root_1])
    elif desc>0: 
        return ([root_2, root_1])
    else:
        return []
    
zero_grad(1, 0, 0, 0)
```

```python
def min_point(a, b, c, d, e, f):
    x = (e*b-2*c*d)/(4*a*c-b**2)
    y = (b*d-2*a*e)/(4*a*c-b**2)
    return (x,y)

min_point(2, 1, 1, -5, -3, 57)
```

```python
import numpy as np
def gradient(a, b, c, d, e, f, x0, y0, rate):

    x, y = x0, y0
    step_x = 2*a*x+b*y+d
    step_y = b*x+2*c*y+e
    lenght = np.sqrt(np.square(step_x)+np.square(step_y))
    x = x - rate/lenght*step_x
    y = y - rate/lenght*step_y
    return (x,y)

gradient(2, 1, 1, -5, -3, 57, 0, 0, 1)
```

```python
def  min_point(f,a,b):
    while b-a>0.0001:
        t1 = a+(b-a)/3
        t2 = a+2*(b-a)/3
        f_a = f(a)
        f_b = f(b)
        f_t1 = f(t1)
        f_t2 = f(t2)
        if min(f_a, f_t1) <= min(f_t2, f_b):
            b = t2
        else:
            a = t1
    return a

min_point(lambda x: (x-1)**2, -10, 10)
```
