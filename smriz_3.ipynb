# 3. **Формальная математическая модель**

## **Для варианта постановки задачи 1**
$X_1,X_2,X_3$ - количество тонн определенной шихты/смеси для выплавки стали

**Ограничения на количество материалов:**

$X_{1}+X_{2}+X_{3}\leq 200$

$X_{1}\leq 150$

$X_{2}\leq 100$

$X_{3}\leq 75$

**Дополнительные ограничения:**

$\sum(X) \geq 0$ - неотрицательный вес шихт

**Ограничения по ГОСТу:**

$\frac{(0.15X_1 + 0.15X_2 + 0.17X_3)}{X_{1}+X_{2}+X_{3}} \geq 0.16$

$\frac{(0.15X_100*0.15X_2 + 0.17X_3)}{X_{1}+X_{2}+X_{3}} \leq 0.18$

$\frac{(0.1X_1 + 0.08X_2 + 0.09X_3)}{X_{1}+X_{2}+X_{3}} \leq 0.09$

$\frac{(0.007X_1 + 0.003X_2 + 0.005X_3)}{X_{1}+X_{2}+X_{3}} \leq 0.005$

**Ограничение по стоимости стоимости:**

В данной задаче нужно максимально загрузить печь при минимальной стоимости шихт. За минимальную стоимость приму результат, полученный в задаче $2-это 150 \$$ за тонну. Но в данной задаче у нас есть $200$ тонная печь. А значит минимальная стоимость при максимальной загрузке такой печи можно условно оценикть, как $150*200=30000\$$. 

$(Price*x)<=30000$

**Целевая функция (максимизация загрузки):**

$\sum(X)\rightarrow \max$

#### **Вариант_1**

```
import numpy as np
import cvxpy
C = np.array([[0.15,0.10,0.007],[0.15,0.08,0.003],[0.17,0.09,0.005]]) #проценты записаны в долях
Price = np.array([300,200,150])#цена за тонну
Weight = np.array([[150,100,75]])#тонн доступно
x = cvxpy.Variable(shape=(1,C.shape[0]), integer = False)
total_cost = cvxpy.sum(x,axis=1)
C0 = x>=0
C1 = cvxpy.sum(x,axis=1)<=200
C2 = x*C[:,0] >=0.16*cvxpy.sum(x,axis=1)
C3 = x*C[:,0] <=0.18*cvxpy.sum(x,axis=1)
C4 = x*C[:,1] <=0.09*cvxpy.sum(x,axis=1)
C5 = x*C[:,2] <=0.005*cvxpy.sum(x,axis=1)
C6 = x[:,0]<= 150
C7 = x[:,1]<= 100
C8 = x[:,2]<= 75
C9 = x*Price<=200*150
problem = cvxpy.Problem(cvxpy.Maximize(total_cost), constraints=[C0,C1,C2,C3,C4,C5,C6,C7,C8,C9])
print('Загрузка печи:',problem.solve(),'\nСостав:',np.round(x.value,3),'\nСтоимость:',np.sum(x.value*Price))
print('Количество хрома в итоговой смеси:{} тонн,{} % от всей смеси'.format(round(np.sum(C[:,0]*x.value),2),round(np.sum(C[:,0]*x.value)/np.sum(x.value)*100,2)))
print('Количество никеля в итоговой смеси:{} тонн,{} % от всей смеси'.format(round(np.sum(C[:,1]*x.value),2),round(np.sum(C[:,1]*x.value)/np.sum(x.value)*100,2)))
print('Количество фосфора в итоговой смеси:{} тонн,{} % от всей смеси'.format(round(np.sum(C[:,2]*x.value),2),round(np.sum(C[:,2]*x.value)/np.sum(x.value)*100,2)))
```

```
import numpy as np
import cvxpy

C = np.array([[0.15,0.10,0.007],[0.15,0.08,0.003],[0.17,0.09,0.005]])

Weight = np.array([150,100,75])#тонн доступно
Price = np.array([300,200,150])

x = cvxpy.Variable(shape=(1,C.shape[0]), integer = False)
total_weight = cvxpy.sum(x)
C0 = x>=0
C1 = cvxpy.sum(x,axis=1)<=200
C2 = x*C[:,0] >=0.16*cvxpy.sum(x,axis=1)
C3 = x*C[:,0] <=0.18*cvxpy.sum(x,axis=1)
C4 = x*C[:,1]<=0.09*cvxpy.sum(x,axis=1)
C5 = x*C[:,2]<=0.005*cvxpy.sum(x,axis=1)
C6 = x[:,0]<= 150
C7 = x[:,1]<= 100
C8 = x[:,2]<= 75

problem = cvxpy.Problem(cvxpy.Maximize(total_weight), constraints=[C0,C1,C2,C3,C4,C5,C6,C7,C8])
print('Максимальный вес загрузки:',problem.solve())

total_cost = x*Price
C1 = cvxpy.sum(x,axis=1)==150 # Используем суммарный максимальновозможный вес загрузки, как ограничение типа равенство


problem = cvxpy.Problem(cvxpy.Minimize(total_cost), constraints=[C0,C1,C2,C3,C4,C5,C6,C7,C8],)
print('Доли каждой шихты:',problem.solve(),'\nСостав:',np.round(x.value))

print('Количество хрома в итоговой смеси:{} тонн,{} % от всей смеси'.format(round(np.sum(C[:,0]*x.value),2),round(np.sum(C[:,0]*x.value)/np.sum(x.value)*100,2)))
print('Количество никеля в итоговой смеси:{} тонн,{} % от всей смеси'.format(round(np.sum(C[:,1]*x.value),2),round(np.sum(C[:,1]*x.value)/np.sum(x.value)*100,2)))
print('Количество фосфора в итоговой смеси:{} тонн,{} % от всей смеси'.format(round(np.sum(C[:,2]*x.value),2),round(np.sum(C[:,2]*x.value)/np.sum(x.value)*100,2)))
```
