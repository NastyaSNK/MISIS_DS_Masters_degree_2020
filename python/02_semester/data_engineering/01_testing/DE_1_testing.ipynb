```python
def check_server(mode):
    if mode == 'memory':
        return 'Memory is ok'
    elif mode == 'connection':
        return 'Connection is ok'
    else:
         raise ValueError()
            
check_server('memory')
```

```python
def safe_sum(x,y):
    try:
        return (x+y)
    except TypeError:
        print('Can\'t sum {} and {}'.format(x,y))
        return (0)
safe_sum(5, 'a')
```

```python
from copy import copy

def remove_dups(values):
    values2 = list(reversed(values))
    for i in range(len(values)):
        if values[i] in values[i+1:]:
            values2.remove(values[i])
    return list(reversed(values2))

remove_dups([1, 12, 4, 1, 4, 8])
```

```python
from collections import Counter  
import pdb  
import pandas as pd
  
imdb_data = pd.read_csv(r'C:\Users\snkna\Downloads\imdb.xls')
def count_genres(column):  
    genres = []  
    # ставим брейкпоинт в этом месте  
    pdb.set_trace()  
    for movie_genres in column:  
        splitted = movie_genres.split(",")  
        genres.extend(splitted)  
    counter = Counter(genres)  
      
    return counter   
   
print(count_genres(imdb_data["Genre"]))  
```

```python
from collections import defaultdict


def group_values(db, value_key, group_key, step):
    grouped = defaultdict(list) 
    for item in db:
        grouped[(item[group_key] // step)*step].append(item[value_key])
    return grouped
```
