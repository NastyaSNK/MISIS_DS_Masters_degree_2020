# Задание 15.2

Используйте функцию ```is_prime()```.

Выведите все простые числа от $2$ до $n$ включительно. Код, определяющий простоту числа, вынесите в отдельную функцию ```is_prime()```.

***Подсказка*** *: Число называется простым, если делится только на $1$ и само на себя.*

````python
def is_prime(n):
    count = 0
    for i in range(2, n):
        if n % i == 0:
            count += 1
    return True if count == 0 else False


for i in range(2, n + 1):
    if is_prime(i):
        print(i)
````
